# Bitcoin Exploratory Regression Analysis

This notebook investigates the price of Bitcoin over time.

## Load data

```python
from bitchaser.data.load import (
    load_latest,
    load_snapshot,
)

btc_snapshot = load_snapshot()
btc_latest = load_latest()

print(f"Loaded {len(btc_snapshot):,} snapshot observations.")
print(f"Loaded {len(btc_latest):,} latest observations.")
```

```python
# Plot settings
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (12, 6)
plt.rcParams["axes.grid"] = True
```

## Plot BTC/USD

```python
display(btc_snapshot.head())
display(btc_snapshot.tail())
```

### BTC/USD vs Time

```python
ax = btc_snapshot.plot(
    x="date",
    y="close",
    title="Bitcoin Daily Closing Price",
    xlabel="Year",
    ylabel="BTC/USD",
    legend=False,
)
```

### Logarithmic BTC/USD vs Time

```python
ax = btc_snapshot.plot(
    x="date",
    y="close",
    logy=True,
    title="Bitcoin Daily Closing Price",
    xlabel="Year",
    ylabel="BTC/USD",
    legend=False,
)
```

### Logarithmic BTC/USD vs Logarithmic Time

```python
ax = btc_snapshot.plot(
    x="days",
    y="close",
    logy=True,
    logx=True,
    title="Bitcoin Daily Closing Price",
    xlabel="DABL",
    ylabel="BTC/USD",
    legend=False,
)
```
